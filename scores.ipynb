- **Importing the modules**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Get your data
data = pd.read_csv('students_scores.csv')

# Call your data
data
```

- **Part-B Practical(Python Programming)**

Step 1: Measures of Central Tendency & Dispersion

• Calculate mean, median, mode of Math_ Score.

• Find range, variance, standard deviation of Science_Score.

```python
# Calculate the mean of Math_Score
mean_math_score = data['Math_Score'].mean()

# Calculate the median of Math_Score
mean_math_score = data['Math_Score'].median()

# Calculate the mode of Math_Score
mean_math_score = data['Math_Score'].mode()

# Print the results
print(f"Mean: {mean_math_score}")
print(f"Median: {mean_math_score}")
print(f"Mode: {mean_math_score}")

print("\n------------------------------------------------------\n")

# Calculate the range of the Science_Score
range_science_score = data['Science_Score'].max() - data['Science_Score'].min()

# Calculate the variance of the Science_score
variance_science_scor = data['Science_Score'].var()

# Calculate the standard deviation of the Science_score
std_dev_science_score = data['Science_Score'].std()

# Print the results
print(f"Range: {range_science_score}")
print(f"Variance: {variance_science_scor}")
print(f"Standard Deviation: {std_dev_science_score}")
```

Step 2: Probability Basics

• Find the probability of students passing (Pass_Fail = 1).

• Create a contingency table between Pass_Fail and Hours_Studied > 5.

• Calculate conditional probability: P(Pass | Hours_Studied > 5).

```python
# Probability of Student Passing
prob_pass = (data['Pass_Fail']== 1).mean()
print(f"Probability of a student passing: {prob_pass}")

# Create a contingency table

data['studied_5_hours'] = data['Hours_Studied'] > 5
contingency_table = pd.crosstab(data['Pass_Fail'], data['studied_5_hours'])
print(contingency_table)

# Calculate the conditional Probability

contingency_table = pd.crosstab(data['Pass_Fail'], data['Hours_Studied'] > 5)
p_pass_given_gt5 = contingency_table.loc[1, True] / contingency_table[True].sum()
print("Conditional Probability P(Pass | Hours_Studied > 5):", p_pass_given_gt5)
```

Step 3: Distribution & Visualization

• Plot a Histogram & Normal curve for Math Score.

• Calculate Skewness & Kurtosis for Science Score.

• Perform a Q-Q Plot for English_Score.

```python
# Creating the histogram for Math_Score
plt.hist(data['Math_Score'], bins=10, edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Number of Students')
plt.show()

# Creating a normal curve for math scores
sns.kdeplot(data['Math_Score'])
plt.title('Normal Curve of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Density')
plt.show()

# Calculating the skewness and kurtosis for Math_Score
skewness = data['Math_Score'].skew()
kurtosis = data['Math_Score'].kurtosis()
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")

# Performing a Q-Q plot for English_Score
stats.probplot(data['English_Score'], dist="norm", plot=plt)
plt.title('Q-Q Plot for English Scores')
plt.show()
```

Step 4: Linear Algebra Mini Task

• Represent Math_Score and Science_Score of first 5 students as vectors.

- Perform:
    - Dot product of the two vectors.
    - Norm 1 and Norm 2 of Math_ Score vector.
    - Find angle between the two vectors.

```python
# Represting Math_Score and Science_Score for the first 5 students as vectors
math_scores = data['Math_Score'].head().values
science_scores = data['Science_Score'].head().values
print(f"Math Scores Vector: {math_scores}")
print(f"Science Scores Vector: {science_scores}")

# Dot product of the two vectors
dot_product = np.dot(math_scores, science_scores)
print(f"Dot Product of Math and Science Scores: {dot_product}")

# Norms of Math_Score vector

# Norm 1(Manhattan norm(L1))
norm_math_l1 = np.linalg.norm(math_scores, ord=1)
print(f"Norm of Math Scores Vector: {norm_math_l1}")

# Norm 2(Euclidean norm(L2))
norm_math_l2 = np.linalg.norm(math_scores)
print(f"Norm of Math Scores Vector: {norm_math_l2}")

# Finding the angle between the two vectors
magnitude_math = np.linalg.norm(math_scores)
magnitude_science = np.linalg.norm(science_scores)
cos_theta = dot_product / (magnitude_math * magnitude_science)
angle = np.arccos(cos_theta) * (180/np.pi)
print(f"Angle between Math and Science Scores Vectors: {angle} degrees")
```
